```python
import numpy as np
class VectorCartesiano:


    """Iniciamos el costructor"""
    def __init__(self, x,y,z):
        self.x=float(x)
        self.y=float(y)
        self.z=float(z)
        self.vector=[x,y,z]
        self.mag=((self.x**2) +( self.y**2) + (self.z**2))**0.5 #magnitud
        
        
    def __add__(self,other):
        '''Sobrecarga del operador suma'''
        return VectorCartesiano(self.x + other.x, self.y + other.y, self.z + other.z)
           
    def __sub__(self,other):        
        '''Sobrecarga del operador resta'''
        return VectorCartesiano(self.x - other.x, self.y - other.y, self.z - other.z)
    
    def __mul__(self,other):
        '''Sobrecarga del operador multiplicacion para hacer el producto interno'''
        return (self.x * other.x + self.y * other.y + self.z * other.z) 
    
    def Cruz(self,other):
        """Creamos el metodo para el producto cruz"""
        return VectorCartesiano((self.y * other.z) - (self.z * other.y ) , -((self.x * other.z) - (self.z * other.x )), (self.x * other.y) - (self.y * other.x ))
    
    def Print(self):
        '''Imprime el vector'''
        print(f"[{self.x},{self.y},{self.z}]")
        
    def __getitem__(self,i):
        """Esta operador me imprime la coordenada que se desea"""
        print ( self.vector[i])
  
    def __eq__(self, other):
        """Este operador nos permite saber si dos vectores son iguales"""
        if (self.x==other.x and self.y==other.y and self.z==other.z):
            print(" son iguales uwu")
        else:
            print (" son distintos u.u")
    
    def magnitud(self):
        """Este metodo me da la magnitud del vector"""
        
        print (self.mag)
    

        
        
       

```

```python
class VectorPolar(VectorCartesiano):
    """transforma un vector con coordenadas cartesianas a esfericas"""

    def __init__(self, x,y,z): # constructor coordenadas esf√©ricas
        
        VectorCartesiano.__init__(self,x,y,z) #unpack

        #----estas son las coordenadas
        self.r=self.mag
        self.teta=np.arccos(self.z/self.r)
        if self.x==0:
            self.phi=0
        else:
            self.phi=np.arctan(self.y/self.x)
        
        if self.r<0:
            print("la coordenada r esta fuera del rango, introducir valores de r distintos de 0")
        if np.pi < self.teta or self.teta<0:
            print("la coordenada theta esta fuera del rango")
        if 2 * np.pi < self.phi  or self.phi <0 :
            print("la coordenada phi esta fuera del rango")
        
        

        #----Ahora para los vectores unitarios:
        
        self.ru= np.sin(self.teta)*np.cos(self.phi) + np.sin(self.teta)*np.sin(self.phi) + np.cos(self.teta) #r gorrito
        self.tetau= np.cos(self.teta)*np.cos(self.phi) + np.cos(self.teta)*np.sin(self.phi) - np.sin(self.teta) #teta gorrito
        self.phiu= -np.sin(self.phi)+ np.cos(self.teta) #phi gorrito
    
        self.rr=self.r*self.ru
        self.tt=self.teta*self.tetau
        self.pp=self.phi*self.phiu
    
    def __add__(self,other):
        '''Sobrecarga del operador suma'''
        return VectorCartesiano(self.rr + other.rr, self.tt + other.tt, self.pp + other.pp)
           
    def __sub__(self,other):        
        '''Sobrecarga del operador resta'''
        return VectorCartesiano(self.rr - other.rr, self.tt - other.tt, self.pp - other.pp)
    
    def __mul__(self,other):
        '''Sobrecarga del operador multiplicacion para hacer el producto interno'''
        return (self.rr * other.rr + self.tt * other.tt + self.pp * other.pp) 
    
    def Cruz(self,other):
        """Creamos el metodo para el producto cruz"""
        return VectorCartesiano((self.tt * other.pp) - (self.pp * other.tt ) , -((self.rr * other.pp) - (self.pp * other.rr )), (self.rr * other.tt) - (self.tt * other.rr ))

        
    def __getitem__(self,i):
        """Esta operador me imprime la coordenada que se desea"""
        print ( self.vector[i])
  
    def __eq__(self, other):
        """Este operador nos permite saber si dos vectores son iguales"""
        if (self.rr==other.rr and self.tt==other.tt and self.pp==other.pp):
            print(" son iguales uwu")
        else:
            print (" son distintos u.u")
    
    def magnitud(self):
        """Este metodo me da la magnitud del vector"""
        
        print (self.mag)
    def Printpol(self):
        '''Imprime el vector en coordenadas esfericas'''
        
        print(f"[{self.rr},{self.tt},{self.pp}]")

vecpol1=VectorPolar(1,1,1)
print ("vecpol1= ", end=" ")
vecpol1.Printpol()
print ("vecpol2= ", end=" ")
vecpol2=VectorPolar(2,5,2)
vecpol2.Printpol()
print("la magnitud del vec1 en coordenadas esfericas es: ",vecpol1.mag , "podemos ver que permanece constante")  #notese que vec1 = vecpol1

vecc=vecpol1+vecpol2
vecc.Print()
```
